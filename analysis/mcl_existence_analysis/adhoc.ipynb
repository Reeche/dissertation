```python
import matplotlib.pyplot as plt

exp_adap = [3, 15]
exp_mal = [18, 9]
control_adapt = [5, 5]
control_mal = [15, 11]

x = ["Pre-training", "Post-training"]


# plot those points and add lines in between the points
plt.plot(x, exp_mal, 'ro-', label='Experimental maladaptive', markersize=10)
plt.plot(x, exp_adap, 'bo-', label='Experimental adaptive', markersize=10)
plt.plot(x, control_mal, 'rv-', label='Control maladaptive', markersize=10)
plt.plot(x, control_adapt, 'bv-', label='Control adaptive', markersize=10)


plt.legend()
plt.ylabel('Count of strategies')
plt.savefig('strategy_counts.png')
#plt.show()
```
