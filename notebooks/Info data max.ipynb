```python
import numpy as np
import pandas as pd
from learning_utils import pickle_load
```

```python
result_dir = "results/info_v1.0_data_max"
import os
last_k = 50
fs = os.listdir(result_dir)
total_data = []
for f in fs:
    b = f.split("_")
    pid = int(b[0])
    model = int(b[3].split(".")[0])
    data = pickle_load(f"{result_dir}/{f}")
    info = data['info']
    for sim_info in info:
        total_sims = []
        for t_info in sim_info[-last_k:]:
            for t in t_info:
                total_sims.append([*t, model])
        total_data += total_sims
print(len(total_data))
```

```python
data_df = pd.DataFrame(total_data, columns=['info', 'node_value', 'pr', 'rpe', 'model'])
```

```python
from IPython.display import display
display(data_df)
```

```python
pd.options.display.max_rows = 10000
data_df['info_present'] =  ~data_df['info']
data_df['positive_value'] = data_df['node_value'] > 0
display(data_df)
```

```python
grouped_data_df = data_df.groupby(['info_present', 'positive_value', 'model'])
display(grouped_data_df)
```

```python
mean_data_df = grouped_data_df.mean().reset_index()
display(mean_data_df)
```
